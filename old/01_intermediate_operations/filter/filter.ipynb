# Filter

The filter operation is used to selectively include or exclude elements from a stream based on a given condition.

<br>
<center></center>


## Example

Filter even numbers from a given list.

```java
import java.util.stream.*;
```

```java
IntStream
    .range(1, 10)
    .filter(n -> n % 2 == 0)
    .forEach(System.out::println);
```
